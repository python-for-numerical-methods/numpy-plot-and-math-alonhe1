אלגברה ליניארית
נניח מערכת של 3 משוואות בשלושה משתנים:

2x+3y+z=11

x+2y+3z=13

3x+y+2z=12

אפשר לכתוב את זה בצורת

Ax=b

מטריצת מקדמים

A=

[2 3 1

1 2 3

3 1 2]

וקטור הנעלמים

x=

[x

y

z]

וקטור המקדמים
b=

[11

13

12]

````python
# מטריצת המקדמים A
A = np.array([[2, 3, 1],
              [1, 2, 3],
              [3, 1, 2]])
# וקטור התוצאה b
b = np.array([11, 13, 12])
# פתרון המערכת
x = np.linalg.solve(A, b)
print("פתרון")
print("x =", x[0])  # המיקום הראשון בווקטור x
print("y =", x[1])  # המיקום השני בווקטור x
print("z =", x[2])  # המיקום השלישי בווקטור x
````

# $\frac{\sin{x}}{x}$

````python
xs_with_zero = np.linspace(-10,10,100)
xs = []
for x in xs_with_zero:
    if x != 0:
        xs.append(x)

ys = []
for x in xs:
    ys.append(math.sin(x)/x)

plt.plot(xs,ys,marker='o')
plt.show()
````

# $ x \sin{\frac{1}{x}} $

````python
xs_with_zero = np.linspace(-0.1,0.1,1000)
xs = []
for x in xs_with_zero:
    if x != 0:
        xs.append(x)


ys = []
for x in xs:
    ys.append(x * math.sin(1/x))

plt.plot(xs,ys,marker='o')
plt.show()
````

# תרגיל בית
יש לכתוב פונקציה בשם normalize_array שמקבלת מערך NumPy חד‑ממדי של מספרים ממשיים, ומחזירה מערך חדש מנורמל בתחום
[0,1]

לפי הנוסחה:

x_norm = x−min(x)/(max(x)−min(x))

הקלט הוא מערך numpy חד־ממדי

אם כל הערכים שווים (למשל [5, 5, 5]), החזירו מערך של אפסים באותו גודל (כדי להימנע מחלוקה באפס).

הפתרון חייב להשתמש בפעולות וקטוריות של NumPy בלבד, ללא לולאות for/while.

הפונקציה לא משנה את הקלט המקורי, אלא מחזירה עותק חדש.


````python
import numpy as np

def normalized_array(input_array):
    min_val = np.min(input_array)
    max_val = np.max(input_array)

    if max_val == min_val:
        return np.zeros_like(input_array, dtype=float)

    new_array = (input_array - min_val) / (max_val - min_val)
    return new_array
````
